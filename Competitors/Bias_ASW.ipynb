```python
import pandas as pd
```

```python
df = pd.read_json('posts.json')
df
```

```python
corpus=df["descripcion"].values.tolist()
print(corpus)
```

```python
import collections
import re

# Unimos todos los strings del corpus en uno solo
corpus = ' '.join(corpus)

# Convertimos el string a minÃºsculas para que no haya distinciÃ³n entre mayÃºsculas y minÃºsculas
corpus = corpus.lower()

# Separamos el string en palabras utilizando cualquier caracter que no sea una letra o un nÃºmero como separador
palabras = re.findall(r'\b\w+\b', corpus)

# Contamos la frecuencia de cada palabra y las ordenamos por frecuencia descendente
frecuencia_palabras = collections.Counter(palabras).most_common()

print(frecuencia_palabras)
```

```python
# Calculamos el nÃºmero total de palabras en el corpus
num_palabras = len(palabras)

# Creamos un diccionario con la frecuencia de cada palabra y su porcentaje en relaciÃ³n al nÃºmero total de palabras
frecuencia_porcentaje = {palabra: (frecuencia/num_palabras)*100 for palabra, frecuencia in frecuencia_palabras}

# Imprimimos el diccionario ordenado por frecuencia descendente
for palabra, frecuencia in sorted(frecuencia_porcentaje.items(), key=lambda item: item[1], reverse=True):
    print(palabra, ": ", round(frecuencia, 2), "%")
```

```python
words = ['algorithm', 'bias', 'big data', 'coding', 'computer', 'cybersecurity', 'data', 'digital divide', 'discrimination', 'ethics', 'gender', 'internet', 'machine learning', 'privacy', 'programming', 'robot', 'software', 'technology']
bias_words = ['ageism', 'ableism', 'racism', 'sexism', 'heterosexism', 'homophobia', 'transphobia', 'colorism', 'lookism', 'sizeism', 'classism', 'xenophobia']
words2 = ['Manpower', 'Chairman', 'Businessman', 'Freshman', 'Journeyman', 'Salesman', 'Mankind', 'Right-hand man', 'Workmanlike', 'Policeman', 'Fireman', 'Postman', 'Forefathers', 'Craftsman', 'Repairman', 'Mansplain', 'He', "She","him","her", 'Artificial Intelligence']
total_words=list(set(words+bias_words+words2))
total_words
```

```python
# Contamos la frecuencia de cada palabra en la lista total_words
frecuencia_total_words = {palabra: palabras.count(palabra) for palabra in total_words}

# Ordenamos el diccionario por el valor (frecuencia) en orden descendente
frecuencia_total_words = dict(sorted(frecuencia_total_words.items(), key=lambda item: item[1], reverse=True))

print(frecuencia_total_words)
```
